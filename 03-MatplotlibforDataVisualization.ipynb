```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Import the Dataset

df = pd.read_csv('D:\DataSets\sales_data.csv')
```

```python
df.head()
```

```python
# Basic Line Plot
plt.plot(df['Date'], df['Total Sales'])
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar Graph for Units Sold by Product
df.groupby('Product')['Units Sold'].sum().plot(kind='bar')
plt.title('Total Units Sold by Product')
plt.xlabel('Product')
plt.ylabel('Total Units Sold')
plt.show()
```

```python
# Pie Chart of Total Sales by Region
df.groupby('Region')['Total Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Total Sales by Region')
plt.ylabel('')  # Hide the y-label to avoid clutter
plt.show()
```

```python
# Scatter Plot of Unit Price vs. Units Sold
plt.scatter(df['Unit Price'], df['Units Sold'])
plt.title('Unit Price vs. Units Sold')
plt.xlabel('Unit Price')
plt.ylabel('Units Sold')
plt.show()
```

```python
# Line Plot with Multiple Lines for different Products
for product, product_data in df.groupby('Product'):
    plt.plot(product_data['Date'], product_data['Total Sales'], label=product)
plt.title('Total Sales Over Time by Product')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Compute the correlation matrix
corr = df[['Units Sold', 'Unit Price', 'Total Sales']].corr()

# Plot the heatmap
fig, ax = plt.subplots()
cax = ax.matshow(corr, cmap='coolwarm')
fig.colorbar(cax)
ticks = np.arange(0, len(corr.columns), 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)
plt.title('Correlation Heatmap', pad=20)
plt.show()
```

```python
# Scatter Plot of Unit Price vs. Units Sold
plt.scatter(df['Unit Price'], df['Units Sold'])
plt.title('Unit Price vs. Units Sold')
plt.xlabel('Unit Price')
plt.ylabel('Units Sold')
plt.show()

# Subplots for Total Sales by Product and Region
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Total Sales by Product
sales_by_product = df['Total Sales'].groupby(df['Product']).sum()
axs[0, 0].bar(sales_by_product.index, sales_by_product)
axs[0, 0].set_title('Total Sales by Product')
axs[0, 0].set_xlabel('Product')
axs[0, 0].set_ylabel('Total Sales')

# Total Sales by Region
sales_by_region = df['Total Sales'].groupby(df['Region']).sum()
axs[0, 1].bar(sales_by_region.index, sales_by_region, color='orange')
axs[0, 1].set_title('Total Sales by Region')
axs[0, 1].set_xlabel('Region')
axs[0, 1].set_ylabel('Total Sales')

# Units Sold Over Time
axs[1, 0].plot(df['Date'], df['Units Sold'])
axs[1, 0].set_title('Units Sold Over Time')
axs[1, 0].set_xlabel('Date')
axs[1, 0].set_ylabel('Units Sold')
axs[1, 0].tick_params(axis='x', rotation=45)

# Total Sales Over Time
axs[1, 1].plot(df['Date'], df['Total Sales'], color='green')
axs[1, 1].set_title('Total Sales Over Time')
axs[1, 1].set_xlabel('Date')
axs[1, 1].set_ylabel('Total Sales')
axs[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Box Plot for Units Sold by Product
df.boxplot(column='Units Sold', by='Product')
plt.title('Box Plot of Units Sold by Product')
plt.suptitle('')  # Remove the automatic suptitle
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.show()

# Box Plot for Total Sales by Region
df.boxplot(column='Total Sales', by='Region')
plt.title('Box Plot of Total Sales by Region')
plt.suptitle('')  # Remove the automatic suptitle
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Donut Chart for Total Sales by Region
sales_by_region = df.groupby('Region')['Total Sales'].sum()
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(sales_by_region, labels=sales_by_region.index, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3, edgecolor='w'))
# Draw circle for the center
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')  
plt.title('Total Sales by Region (Donut Chart)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data
np.random.seed(0)
data = pd.DataFrame({
    'x': np.random.normal(loc=0, scale=1, size=100),
    'y': np.random.normal(loc=0, scale=1, size=100),
    'category': np.random.choice(['A', 'B', 'C'], size=100)
})

# Scatter plot
sns.scatterplot(x='x', y='y', data=data)
plt.title('Scatter Plot Example')
plt.show()

# Line plot
sns.lineplot(x=np.arange(100), y=data['y'])
plt.title('Line Plot Example')
plt.show()

# Bar plot
sns.barplot(x='category', y='x', data=data)
plt.title('Bar Plot Example')
plt.show()

# Histogram
sns.histplot(data['y'], kde=True)
plt.title('Histogram Example')
plt.show()

# Box plot
sns.boxplot(x='category', y='y', data=data)
plt.title('Box Plot Example')
plt.show()

# Heatmap
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap Example')
plt.show()
```
